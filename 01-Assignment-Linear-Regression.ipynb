```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Ensure reproducibility
np.random.seed(42)

# Generate pseudo-random data points close to a line y = 2x + 1
x = np.linspace(0, 10, 20)  # 20 evenly spaced points from 0 to 10
y = 2 * x + 1 + np.random.normal(scale=1, size=x.shape)  # Adding noise

# Reshape x for sklearn (expects 2D input)
x = x.reshape(-1, 1)

# Perform linear regression
model = LinearRegression()
model.fit(x, y)

# Extract slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Predict y for x=7
x_sample = np.array([[7]])
y_pred_sample = model.predict(x_sample)

# Plot data points and regression line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, model.predict(x), color='red', label=f'Regression line: y = {slope:.2f}x + {intercept:.2f}')
plt.scatter(x_sample, y_pred_sample, color='green', label=f'Prediction for x=7: y={y_pred_sample[0]:.2f}', marker='x', s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression Demonstration')
plt.show()

# Print results
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"Prediction for x=7: {y_pred_sample[0]:.2f}")
```
