
# Linear Regression with Visualization and Prediction

This notebook demonstrates a **simple linear regression** example using:
- Synthetic data
- Least Squares Linear Regression
- Visualization of data, fitted line, and prediction

You can run this notebook directly in **Google Colab**.


```python

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```


## Step 1: Generate Sample Data

We create a simple linear relationship with noise:

\[
y = 2x + 3 + \text{noise}
\]


```python

# Generate synthetic data
np.random.seed(42)
X = np.linspace(0, 10, 30).reshape(-1, 1)
y = 2 * X.flatten() + 3 + np.random.normal(0, 2, size=X.shape[0])
```


## Step 2: Fit Linear Regression Model


```python

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Model parameters
slope = model.coef_[0]
intercept = model.intercept_

slope, intercept
```


## Step 3: Make Predictions


```python

# Prediction range
X_pred = np.linspace(0, 12, 100).reshape(-1, 1)
y_pred = model.predict(X_pred)

# Predict a single future value
x_new = np.array([[11]])
y_new = model.predict(x_new)

x_new, y_new
```


## Step 4: Visualization

- Blue dots: data points  
- Red line: regression model  
- Green star: predicted point


```python

plt.figure(figsize=(8,5))
plt.scatter(X, y, label="Data points")
plt.plot(X_pred, y_pred, label="Regression line")
plt.scatter(x_new, y_new, marker='*', s=200, label="Prediction")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression with Prediction")
plt.legend()
plt.grid(True)
plt.show()
```


## Interpretation

The regression model learned:
- **Slope**  2.00  
- **Intercept**  3.00  

This means that for every unit increase in `x`, the predicted value of `y`
increases by approximately the slope.
