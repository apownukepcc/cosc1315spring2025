```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x = np.array([1.00, 2.00, 3.00, 4.00, 5.00, 6.00])
yo = np.array([3.679136, 5.648717, 7.442161, 9.204735, 11.19715, 13.44109])

# Solve the quadratic regression problem: y = a*x^2 + b*x + c
coeffs = np.polyfit(x, yo, 2) # degree 2 for quadratic model
a, b, c = coeffs
print("Quadratic regression model:")
print(" y = {:.4f} x^2 + {:.4f} x + {:.4f}".format(a, b, c))

# Compute predicted values using the fitted quadratic model
yp = np.polyval(coeffs, x)

# Calculate the Sum of Squared Errors (SSE)
SSE = np.sum((yo - yp) ** 2)
print("Sum of Squared Errors (SSE):", SSE)

# Plot the data points and the quadratic regression curve
plt.figure(figsize=(8,6))
plt.scatter(x, yo, color='blue', label='Data points')
x_curve = np.linspace(min(x), max(x), 100)
y_curve = np.polyval(coeffs, x_curve)
plt.plot(x_curve, y_curve, color='red', label='Quadratic fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Regression Fit')
plt.legend()
plt.show()
```
